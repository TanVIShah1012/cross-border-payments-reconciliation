```python

!pip install --quiet sqlalchemy seaborn

import pandas as pd
import numpy as np
import sqlite3
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

 synthetic transactions CSV

```python
# Parameters
N = 5000  # number of transactions to synthesize
rng = np.random.default_rng(seed=42)

# Helper lists
countries = ['US','GB','CA','MX','BR','AR','IN','PH','ID','NG']
currencies = {'US':'USD','GB':'GBP','CA':'CAD','MX':'MXN','BR':'BRL','AR':'ARS','IN':'INR','PH':'PHP','ID':'IDR','NG':'NGN'}
platform_spread_mean = 0.005  # 0.5% spread typical
platform_spread_sd = 0.002

# Create base columns
transaction_id = [f"tx_{i+1}" for i in range(N)]
country = rng.choice(countries, size=N, p=[0.25,0.12,0.10,0.10,0.08,0.05,0.10,0.08,0.07,0.05])
payout_currency = [currencies[c] for c in country]

# random gross USD amounts with heavy tail
gross_amount_usd = np.round(np.exp(rng.normal(np.log(50), 1.0, size=N)),2)  # median around $50

# fees as percent of gross (random)
fee_pct = np.clip(rng.normal(0.02, 0.01, size=N), 0.002, 0.10)  # 0.2% - 10%
fee_amount_usd = np.round(gross_amount_usd * fee_pct,2)

# platform spread pct small around mean
platform_spread_pct = np.clip(rng.normal(platform_spread_mean, platform_spread_sd, size=N), 0.0005, 0.02)

# fx_rate: simulate typical USD->local rates (approx)
fx_map = {'USD':1.0,'GBP':0.77,'CAD':1.34,'MXN':17.0,'BRL':5.0,'ARS':350.0,'INR':83.0,'PHP':55.0,'IDR':15000.0,'NGN':750.0}
fx_rate = [fx_map[c] for c in payout_currency]

# expected payout local = (gross - fees) * fx_rate * (1 - platform_spread?)
# For simplicity assume expected = (gross - fee) * fx_rate
expected_payout_local = np.round((gross_amount_usd - fee_amount_usd) * fx_rate,2)

# actual payout local adds small random noise and occasional larger discrepancies
noise = rng.normal(0, 2.0, size=N)  # baseline small cents/units noise
# randomly inject real mismatches
mismatch_mask = rng.random(size=N) < 0.02  # 2% true mismatches
larger_noise = rng.normal(0, 200.0, size=N) * mismatch_mask  # larger differences for mismatches
actual_payout_local = np.round(expected_payout_local + noise + larger_noise,2)

# settled_date across several months
dates = pd.date_range(start="2025-06-01", periods=90).to_pydatetime().tolist()
settled_date = [rng.choice(dates) for _ in range(N)]
settled_month = pd.to_datetime(settled_date).to_period('M').astype(str)

# build DataFrame
df = pd.DataFrame({
    'transaction_id': transaction_id,
    'country': country,
    'payout_currency': payout_currency,
    'gross_amount_usd': gross_amount_usd,
    'fee_amount_usd': fee_amount_usd,
    'platform_spread_pct': np.round(platform_spread_pct,6),
    'fx_rate': fx_rate,
    'expected_payout_local': expected_payout_local,
    'actual_payout_local': actual_payout_local,
    'settled_date': settled_date,
    'settled_month': settled_month
})

CSV_PATH = "/content/synthetic_transactions.csv"
df.to_csv(CSV_PATH, index=False)
print("Synthetic CSV written to:", CSV_PATH)
display(df.head())
print("Shape:", df.shape)
```

checks and cleaning


```python

df = pd.read_csv(CSV_PATH)
df.columns = df.columns.str.strip().str.lower()
# convert dates
df['settled_date'] = pd.to_datetime(df['settled_date'], errors='coerce')
df['settled_month'] = df['settled_date'].dt.to_period('M').astype(str)
print("Columns:", df.columns.tolist())
print("Missing values (top 10):")
display(df.isna().sum().sort_values(ascending=False).head(10))
df.info()
```

SQLite DB and import table

```python
DB_PATH = "/content/mural_pay_demo.db"
conn = sqlite3.connect(DB_PATH)
df.to_sql("transactions", conn, if_exists="replace", index=False)
print("Imported to SQLite DB:", DB_PATH)
cur = conn.cursor()
cur.execute("SELECT COUNT(*) FROM transactions")
print("Rows in transactions table:", cur.fetchone()[0])
```

SQL views

```python
cur = conn.cursor()
sql = """
DROP VIEW IF EXISTS reconciliation_view;
CREATE VIEW reconciliation_view AS
SELECT
    transaction_id,
    country,
    payout_currency,
    gross_amount_usd,
    fee_amount_usd,
    expected_payout_local,
    actual_payout_local,
    ROUND(actual_payout_local - expected_payout_local, 2) AS payout_diff_local,
    CASE
        WHEN ABS(actual_payout_local - expected_payout_local) > 50 THEN 'Mismatch'
        ELSE 'OK'
    END AS reconciliation_status
FROM transactions;

DROP VIEW IF EXISTS monthly_kpi;
CREATE VIEW monthly_kpi AS
SELECT
    settled_month,
    COUNT(*) AS total_transactions,
    ROUND(SUM(gross_amount_usd),2) AS total_gross_usd,
    ROUND(SUM(fee_amount_usd),2) AS total_fees_usd,
    ROUND(100.0 * SUM(CASE WHEN ABS(actual_payout_local - expected_payout_local) > 50 THEN 1 ELSE 0 END) / COUNT(*), 2) AS mismatch_rate_pct
FROM transactions
GROUP BY settled_month
ORDER BY settled_month;

DROP VIEW IF EXISTS fx_revenue;
CREATE VIEW fx_revenue AS
SELECT
    payout_currency,
    COUNT(*) AS total_tx,
    ROUND(SUM(gross_amount_usd * COALESCE(platform_spread_pct,0)),2) AS platform_revenue_usd,
    ROUND(AVG(COALESCE(fx_rate,1)),6) AS avg_fx_rate
FROM transactions
GROUP BY payout_currency
ORDER BY platform_revenue_usd DESC;
"""
cur.executescript(sql)
conn.commit()
print("Views created.")
```

Query views into pandas and inspect

```python
mismatches = pd.read_sql("SELECT * FROM reconciliation_view WHERE reconciliation_status='Mismatch' LIMIT 50", conn)
monthly_kpi = pd.read_sql("SELECT * FROM monthly_kpi", conn)
fx_rev = pd.read_sql("SELECT * FROM fx_revenue LIMIT 50", conn)

print("Mismatches sample:")
display(mismatches.head())
print("Monthly KPI:")
display(monthly_kpi)
print("FX Revenue top:")
display(fx_rev)
```

Query views into pandas and inspect

```python
recon_all = pd.read_sql("SELECT transaction_id, payout_diff_local FROM reconciliation_view", conn)
recon_all['payout_diff_local'] = pd.to_numeric(recon_all['payout_diff_local'], errors='coerce')
mean = recon_all['payout_diff_local'].mean()
std = recon_all['payout_diff_local'].std(ddof=0)
recon_all['zscore'] = (recon_all['payout_diff_local'] - mean) / (std if std>0 else 1)
anoms = recon_all[recon_all['zscore'].abs() > 3].sort_values('zscore', ascending=False)
print("Anomalies count (|z|>3):", len(anoms))
display(anoms.head(20))
```

Plots

```python
plt.figure(figsize=(9,4))
sns.barplot(data=monthly_kpi, x='settled_month', y='mismatch_rate_pct')
plt.xticks(rotation=45)
plt.title("Mismatch Rate % by Month")
plt.ylabel("Mismatch Rate %")
plt.xlabel("")
plt.tight_layout()
plt.show()

plt.figure(figsize=(9,4))
vals = recon_all['payout_diff_local'].dropna()
sns.histplot(vals, bins=80)
plt.xlim(np.percentile(vals,1), np.percentile(vals,99))
plt.title("Distribution of payout difference (1-99 pct)")
plt.tight_layout()
plt.show()

plt.figure(figsize=(9,4))
sns.barplot(data=fx_rev.head(10), x='payout_currency', y='platform_revenue_usd')
plt.title("Platform revenue (USD) by payout currency - top 10")
plt.xlabel("")
plt.tight_layout()
plt.show()
```
